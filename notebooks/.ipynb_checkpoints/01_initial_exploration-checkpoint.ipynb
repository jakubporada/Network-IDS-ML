```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("Libraries imported successfully!")
```

```python
# Load Monday data (start with just 10,000 rows)
file_path = '../data/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv'

# Load sample
df_sample = pd.read_csv(file_path, nrows=10000)

print(f"Shape: {df_sample.shape}")
print(f"\nFirst few column names: {df_sample.columns.tolist()[:10]}")
```

```python
# Check the actual column names
print("Column names (first 10):")
print(df_sample.columns.tolist()[:10])
print("\n")
df_sample.head()
```

```python
# What are the labels (attack types)?
print("Label distribution:")
print(df_sample[' Label'].value_counts())  # Note the space before Label
print("\nPercentage:")
print(df_sample[' Label'].value_counts(normalize=True) * 100)
```

```python
# Missing values
print("Missing values per column:")
missing = df_sample.isnull().sum()
print(f"Total missing values: {missing.sum()}")
if missing.sum() > 0:
    print("\nColumns with missing values:")
    print(missing[missing > 0])

# Check for infinity values
print("\n\nChecking for infinity values...")
numeric_cols = df_sample.select_dtypes(include=[np.number]).columns
infinity_found = False
for col in numeric_cols:
    inf_count = np.isinf(df_sample[col]).sum()
    if inf_count > 0:
        print(f"{col}: {inf_count} infinity values")
        infinity_found = True
        
if not infinity_found:
    print("No infinity values found!")
```

```python
# Plot label distribution
plt.figure(figsize=(12, 6))
df_sample[' Label'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title('Traffic Type Distribution (Sample of 10,000 flows)')
plt.xlabel('Traffic Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Calculate percentages
benign_pct = (df_sample[' Label'] == 'BENIGN').sum() / len(df_sample) * 100
attack_pct = 100 - benign_pct
print(f"\nBENIGN: {benign_pct:.1f}%")
print(f"ATTACK: {attack_pct:.1f}%")
```

```python
# Summary
print(f"""
DATA EXPLORATION SUMMARY
========================
Dataset: {file_path.split('/')[-1]}
Rows: {df_sample.shape[0]}
Columns: {df_sample.shape[1]}

Label Distribution:
{df_sample[' Label'].value_counts()}

Key Findings:
- Total missing values: {df_sample.isnull().sum().sum()}
- Benign traffic: {(df_sample[' Label'] == 'BENIGN').sum() / len(df_sample) * 100:.1f}%
- Attack traffic: {(df_sample[' Label'] != 'BENIGN').sum() / len(df_sample) * 100:.1f}%
""")
```
